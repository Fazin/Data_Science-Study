```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')

display(train_set.head())

display(test_set)
```

```
train_set.set_index(keys='NU_INSCRICAO', drop=True)
```

```
new_train = train_set[['NU_INSCRICAO','NU_ANO', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA','NU_IDADE', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'CO_MUNICIPIO_NASCIMENTO', 'NO_MUNICIPIO_NASCIMENTO', 'CO_UF_NASCIMENTO', 'SG_UF_NASCIMENTO', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO', 'CO_ESCOLA', 'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC', 'TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC', 'IN_CERTIFICADO', 'NO_ENTIDADE_CERTIFICACAO', 'CO_UF_ENTIDADE_CERTIFICACAO', 'SG_UF_ENTIDADE_CERTIFICACAO', 'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA', 'CO_UF_PROVA', 'SG_UF_PROVA', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT', 'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT', 'TP_LINGUA', 'TX_GABARITO_CN', 'TX_GABARITO_CH', 'TX_GABARITO_LC', 'TX_GABARITO_MT']
]

new_train.set_index(keys='NU_INSCRICAO')

check1 = new_train[['NU_INSCRICAO','TX_RESPOSTAS_MT','NU_NOTA_MT']]

check1 = check1[check1['NU_NOTA_MT'].notnull()]

check2 = check1['TX_RESPOSTAS_MT'].apply(lambda x: pd.Series(list(x)))

train_fit = pd.concat((check1,check2),join='inner',axis=1)

train_fit.columns = ['NU_INSCRICAO','TX_RESPOSTAS_MT','NU_NOTA_MT','Q_0', 'Q_1', 'Q_2', 'Q_3', 'Q_4', 'Q_5', 'Q_6', 'Q_7', 'Q_8', 'Q_9', 'Q_10', 'Q_11', 'Q_12', 'Q_13', 'Q_14', 'Q_15', 'Q_16', 'Q_17', 'Q_18', 'Q_19', 'Q_20', 'Q_21', 'Q_22', 'Q_23', 'Q_24', 'Q_25', 'Q_26', 'Q_27', 'Q_28', 'Q_29', 'Q_30', 'Q_31', 'Q_32', 'Q_33', 'Q_34', 'Q_35', 'Q_36', 'Q_37', 'Q_38', 'Q_39', 'Q_40', 'Q_41', 'Q_42', 'Q_43', 'Q_44']

display(train_fit.head())

test_1 = test_set[['NU_INSCRICAO','TX_RESPOSTAS_MT']]
test_1 = test_1[test_1['TX_RESPOSTAS_MT'].notnull()]
test_2 = test_1['TX_RESPOSTAS_MT'].apply(lambda x: pd.Series(list(x)))
test_fit = train_fit = pd.concat((test_1,test_2),join='inner',axis=1)
test_fit.columns = ['NU_INSCRICAO','TX_RESPOSTAS_MT','Q_0', 'Q_1', 'Q_2', 'Q_3', 'Q_4', 'Q_5', 'Q_6', 'Q_7', 'Q_8', 'Q_9', 'Q_10', 'Q_11', 'Q_12', 'Q_13', 'Q_14', 'Q_15', 'Q_16', 'Q_17', 'Q_18', 'Q_19', 'Q_20', 'Q_21', 'Q_22', 'Q_23', 'Q_24', 'Q_25', 'Q_26', 'Q_27', 'Q_28', 'Q_29', 'Q_30', 'Q_31', 'Q_32', 'Q_33', 'Q_34', 'Q_35', 'Q_36', 'Q_37', 'Q_38', 'Q_39', 'Q_40', 'Q_41', 'Q_42', 'Q_43', 'Q_44']

display(test_fit.head())
```

```
import seaborn as sns
from statsmodels.formula.api import ols

model = ols('NU_NOTA_MT ~ C(Q_0) + C(Q_1) + C(Q_2) + C(Q_3) + C(Q_4) + C(Q_5) + C(Q_6) + C(Q_7) + C(Q_8) + C(Q_9) + C(Q_10) + C(Q_11) + C(Q_12) + C(Q_13) + C(Q_14) + C(Q_15) + C(Q_16) + C(Q_17) + C(Q_18) + C(Q_19) + C(Q_20) + C(Q_21) + C(Q_22) + C(Q_23) + C(Q_24) + C(Q_25) + C(Q_26) + C(Q_27) + C(Q_28) + C(Q_29) + C(Q_30) + C(Q_31) + C(Q_32) + C(Q_33) + C(Q_34) + C(Q_35) + C(Q_36) + C(Q_37) + C(Q_38) + C(Q_39) + C(Q_40) + C(Q_41) + C(Q_42) + C(Q_43) + C(Q_44)', data=train_fit)
fitted_model = model.fit()
fitted_model.summary()

model.predict()
```
